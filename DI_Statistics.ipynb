```python
df_SM = pd.DataFrame(df[df['Marketing Technique_Social media'] == 1])
df_TA = pd.DataFrame(df[df['Marketing Technique_TV ADs'] == 1])
df_B = pd.DataFrame(df[df['Marketing Technique_Blogs'] == 1])
df_P = pd.DataFrame(df[df['Marketing Technique_Posters'] == 1])
df_PR = pd.DataFrame(df[df['Marketing Technique_Public reviews'] == 1])
df_S = pd.DataFrame(df[df['Marketing Technique_Sponsorship'] == 1])
```

```python
def stats(data):
    data.drop('Marketing Technique_Social media', axis = 'columns', inplace = True)
    data.drop('Marketing Technique_TV ADs', axis = 'columns', inplace = True)
    data.drop('Marketing Technique_Blogs', axis = 'columns', inplace = True)
    data.drop('Marketing Technique_Posters', axis = 'columns', inplace = True)
    data.drop('Marketing Technique_Public reviews', axis = 'columns', inplace = True)
    data.drop('Marketing Technique_Sponsorship', axis = 'columns', inplace = True)
    sample =  pd.get_dummies(data)
    s = sample.sum(axis = 0, skipna = True)
    data1 = pd.array([s["Social Media_Average"], s["Social Media_Least"], s["Social Media_Less"], s["Social Media_More"], s["Social Media_Most"]])
    data2 = pd.array([s["Tele Marketing_Average"], s["Tele Marketing_Least"], s["Tele Marketing_Less"], s["Tele Marketing_More"], s["Tele Marketing_Most"]])
    data3 = pd.array([s["TV Ads_Average"], s["TV Ads_Least"], s["TV Ads_Less"], s["TV Ads_More"], s["TV Ads_Most"]])
    data4 = pd.array([s["Pop-up ADs_Average"], s["Pop-up ADs_Least"], s["Pop-up ADs_Less"], s["Pop-up ADs_More"], s["Pop-up ADs_Most"]])
    data5 = pd.array([s["Posters_Average"], s["Posters_Least"], s["Posters_Less"], s["Posters_More"], s["Posters_Most"]])
    X = np.arange(5)
    data = pd.array(["Average", "Least", "Less", "More", "Most"])
    plt.figure(figsize = (10, 10))
    plt.bar(X + 0.00, data1, label = "Scoial Media", color = 'b', width = 0.10)
    plt.bar(X + 0.15, data2, label = "Tele Marketing", color = 'g', width = 0.10)
    plt.bar(X + 0.30, data3, label = "TV ADs", color = 'r', width = 0.10)
    plt.bar(X + 0.45, data4, label = "Pop-up ADs", color = 'orange', width = 0.10)
    plt.bar(X + 0.60, data5, label = "Posters", color = 'y', width = 0.10)
    plt.xticks(X, data)
    plt.legend()
    plt.show()
```
